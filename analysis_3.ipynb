```python
bar_graph = df['brand'].value_counts().head(25).plot(kind='bar',figsize=(15,5),title='number of products by brands')
bar_graph.setxlabel('brand')
bar_graph.setxylabel('number of products')
```

```python
def clean_price(price):
    if price is None:
        return np.nan
    elif isinstance(price,str):
        price=price.replace(',','')
        if price:
            price=float(price)
        else:
            price=np.nan
        return price
    else:
        return price
df['selling_price'] = df['selling_price'].apply(clean_price)
df['actual_price'] = df['actual_price'].apply(clean_price)
```

```python
df.groupby('category')['selling_price'].mean()
```

```python
seller_data=df.groupby('seller')['selling_price'].mean().reset_index()
seller_data.sort_values('selling_price',ascending=False).head(25)
```
