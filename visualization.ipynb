```python
rainfall = [0.8, 0.6, 0.3, 0.1, 0.0, 1.5, 0.1, 0.3, 0.6, 0.8, 0.9, 0.6]
months = ['jan', 'feb', 'mar', 'apr', 
          'may', 'jun', 'jul', 'aug',
          'sep', 'oct', 'nov', 'dec']
print(len(rainfall), len(months))
fig = plt.figure(figsize=(6,4))
plt.bar(months, rainfall, color='blue', hatch='//')
plt.title('rainfall in 2019')
plt.xlabel('month')
plt.ylabel('rainfall (mm)')
```
